```python
import os, shutil, random

source_path = r"C:\Users\User\PROJECT AKHIR\PROJEK AKHIR SMT 5\archive\Vehicles"
target_path = r"C:\Users\User\PROJECT AKHIR\PROJEK AKHIR SMT 5\archive\dataset"

# pilih kategori yang mau dipakai
classes = ["Bikes","Cars", "Motorcycles", "Trains", "Auto Rickshaws"]  # data disamakan dengan dataset yang ada

# buat folder baru otomatis
for cls in classes:
    os.makedirs(os.path.join(target_path, "train", cls), exist_ok=True)
    os.makedirs(os.path.join(target_path, "test", cls), exist_ok=True)

# split otomatis 80:20
split_ratio = 0.8

for cls in classes:
    files = os.listdir(os.path.join(source_path, cls))
    random.shuffle(files)

    split = int(len(files) * split_ratio)
    train_files = files[:split]
    test_files = files[split:]

    for f in train_files:
        shutil.copy(os.path.join(source_path, cls, f),
                    os.path.join(target_path, "train", cls, f))

    for f in test_files:
        shutil.copy(os.path.join(source_path, cls, f),
                    os.path.join(target_path, "test", cls, f))

print("Dataset berhasil dibuat & dipisah otomatis âœ”")
```
